# Generation of random numbers using module functions and finding moments

1. random() form numpy.random

```python
import numpy.random as rn
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moment

n1=int(input('\nEnter the size for 1D array : '))
x,y=eval(input('Enter the row and column size for 2D array : '))		#int doesn't work here
n=int(input('Enter the size for plot : '))
r=int(input('Enter order of the moment : '))

r1D=rn.random(n1)
r2D=rn.random((x,y))
rplot=rn.random(n)
zcol=np.mean(r2D,axis=0)
zrow=np.mean(r2D,axis=1)
m1D=moment(r1D,moment=r)
m2Dcolumn=moment(r2D,moment=r,axis=0)
m2Drow=moment(r2D,moment=r,axis=1)
	
print()
print('The random() function generates :',rn.random())
print()
print('The random(size=n) function generates :',r1D)
print()
print('The random((x,y)) function generates :',r2D)
print()
print('The moment of order',r,'for the 1D random distribution generated is :',m1D)
print()
print('The moment of order',r,'for the 2D array random distribution generated along the columns is :',m2Dcolumn)
print()
print('The moment of order',r,'for the 2D array random distribution generated along the rows is :',m2Drow)
print()
	
	
plt.figure()
plt.title('Histogram for Random number generation')
plt.hist(rplot,bins=30,rwidth=0.9,range=None,density=False,color='green',label='numpy.random.random()',histtype='bar')
plt.legend(loc='best')
	
plt.figure()
plt.title('Central Limit Theorem')
plt.hist(zcol,bins=30,rwidth=0.9,range=None,density=False,color='blue',label='mean of column terms',histtype='bar')
plt.legend(loc='best')
	
plt.figure()
plt.title('Central Limit Theorem')
plt.hist(zrow,bins=30,rwidth=0.9,range=None,density=False,color='orange',label='mean of row terms',histtype='bar')
plt.legend(loc='best')
	
plt.show()  
```

4. randrange() from random

```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moment

a,b=eval(input('\nEnter lower and upper limit : '))		#int doesn't work here
c=int(input('Enter the step length : '))
rg=r.randrange(a,b,c)
R=int(input('Enter order of the moment : '))
N=int(input('Enter the size of list : '))
List=[r.randrange(a,b,c) for i in range(N)]
m=sum([(i-np.mean(List))**R/len(List) for i in List])
	
print()
print('The randrange() function generates :',rg)
print()
print('The moment of order',R,'for the distribution generated using randrange is :',abs(float(format(m,'0.1f'))))
print()
```

5. gauss() from random

```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moment

mu,sig=eval(input('\nEnter mean and standard deviation : '))		#int doesn't work here
g=r.gauss(mu,sig)
R=int(input('Enter order of the moment : '))
N=int(input('Enter the size of list : '))
List=[r.gauss(mu,sig) for i in range(N)]
m=sum([(i-np.mean(List))**R/len(List) for i in List])
	
print()
print('The gauss() function generates :',g)
print()
print('The moment of order',R,'for the distribution generated using gauss is :',abs(float(format(m,'0.1f'))))
print()
```

9. choice() from numpy.random

```python
import numpy.random as rn
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moment

a=eval(input('Enter the number for numpy.arange ndarray : '))
low,up=eval(input('Enter lower and upper limits of an array : '))
n=int(input('Enter array size : '))
m=int(input('Enter no of elements to be returned : '))
r=int(input('Enter order of the moment : '))
array=np.linspace(low,up,n)
r1=rn.choice(a,size=m,p=None)
r2=rn.choice(array,m)
m1D=moment(r2,moment=r)

print()
print('The choice(a) function generates :',r1)
print()
print('The choice(array) function generates :',r2)
print()
print('The moment of order',r,'for the distribution generated using choice function is :',m1D)
print()

plt.figure()
plt.title('Histogram for Random Number Generation') 
plt.hist(r1,bins=10,rwidth=0.9,color='red')
	
plt.figure()
plt.title('Histogram for Random Number Generation') 
plt.hist(r2,bins=10,rwidth=0.9,color='brown')
	
plt.show()
```
